```python
import pandas

data = pandas.read_csv("datasets/supermarket_data.csv")
data
```

```python
# Use the Describe function
data.describe()
```

```python
# We have noticed that the item price column contains 3 empty slots
# We need to fill these empty slots with a value and in this case we shall fill the slots with the mean of the colum
data["itemprice"] = data["itemprice"].fillna(55)

# Describe the data again
data.describe()
```

```python
# Get the number of entries in a given column
data.groupby("gender").size()
```

```python
# Combining multiple columns
data.groupby(["city", "payment"]).size()
```

```python
# Import the modules
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Below is a count plot showing the distribution of malesto females
sns.countplot(x="gender", data=data, palette="autumn")
plt.title("Gender Distribution Against Total Count")
plt.ylabel("Total Individual Count")
```

```python
# Classwork
sns.countplot(x="product", data=data, palette="dark")
plt.xlabel("Product Category")
plt.ylabel("Total Product Count")
plt.title("Product Distribution Per Category")
plt.xticks (rotation = 45)
```

```python
# Below is a count plot that has 2 columns
sns.countplot(x="gender", data=data, hue="city", palette="autumn")
```

```python
# Below is a pieplot
data["payment"].value_counts().plot.pie(autopct = '%1.3f%%')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# 1. Create the plot and assign it to 'ax'
ax = sns.countplot(x="gender", data=data, palette="spring")

# 2. Add labels to each bar container
for container in ax.containers:
    ax.bar_label(container)

# 3. Rest of your formatting
plt.title("Gender Distribution Against Total Count")
plt.ylabel("Total Count of Individuals")
plt.show()
```

```python
# 1. Get the total count of the dat
total = len(data['gender'])

# 2. Create the plo
ax = sns.countplot(x="gender", data=data, palette="spring")

# 3. Calculate percentages and add them as label
for container in ax.containers:

    # Create labels: (bar height / total) * 10
    labels = [f'{(v/total)*100:0.1f}%' for v in container.datavalues]
    ax.bar_label(container, labels=labels, padding=3)

plt.title("Gender Distribution (Percentage)")
plt.ylabel("Total Count")
plt.show()
```
